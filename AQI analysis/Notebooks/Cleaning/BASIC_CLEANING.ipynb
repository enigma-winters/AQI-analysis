```python
def cleanmaster(pollutant_name,df):
    ndf=pd.DataFrame()
    y=list((Counter(df['station'].values)).keys())#218 stations
    def clean(pollutant_name,df,y):
        temp1,temp2,temp3=[],[],[]
        #############################################################
        for j in range(len(y)):
            temp_df=pd.DataFrame()
            temp_df=df[df['station']==y[j]]
            try:
                temp1.append((temp_df[temp_df['pollutant_id']==pollutant_name].pollutant_min).values[0])  
            except:
                temp1.append(np.nan)
                
            try:    
                temp2.append((temp_df[temp_df['pollutant_id']==pollutant_name].pollutant_max).values[0])
            except:
                temp2.append(np.nan)
            try:    
                temp3.append((temp_df[temp_df['pollutant_id']==pollutant_name].pollutant_avg).values[0])
            except:
                temp3.append(np.nan)
        ##############################################################        
        return temp1,temp2,temp3
    ndf['Station']=y
    ndf[pollutant_name+'_'+'min'],ndf[pollutant_name+'_'+'max'],ndf[pollutant_name+'_'+'avg']=clean(pollutant_name,df,y)        
    return ndf
```

```python
testdf1=cleanmaster('PM10',df)
testdf2=cleanmaster('PM2.5',df)
testdf3=cleanmaster('NO2',df)
testdf4=cleanmaster('NH3',df)
testdf5=cleanmaster('SO2',df)
testdf6=cleanmaster('CO',df)
testdf7=cleanmaster('OZONE',df)
```

```python
result_df=(((((testdf1.set_index('Station').join(testdf2.set_index('Station'))).join(testdf3.set_index('Station'))).join(testdf4.set_index('Station'))).join(testdf5.set_index('Station'))).join(testdf6.set_index('Station'))).join(testdf7.set_index('Station'))
```

```python
date_time=np.array([df['last_update'].values[0]]*len(result_df.index))#add this to final df
```

```python
result_df.insert(0,'Date_time',date_time)
```
